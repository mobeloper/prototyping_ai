### Using Virtual Environments in Python
A virtual environment is an isolated Python environment that allows you to manage dependencies for a specific project independently of other projects. This helps avoid conflicts between package versions across different projects.

#### Why Use a Virtual Environment?
1. **Isolation:** Each project has its own dependencies, avoiding interference with others.
2. **Version Management:** You can maintain different versions of the same library for separate projects.
3. **Reproducibility:** Makes it easier to share your project setup with others using tools like `requirements.txt`.

#### Setting Up a Virtual Environment
**Follow these steps to create and activate a virtual environment:**

##### Windows:
1. Create a virtual environment:
   ```
   python -m venv myenv
   ```
2. Activate the virtual environment:
   ```
   myenv\Scripts\activate
   ```
3. Install dependencies inside the environment:
   ```
   pip install opencv-python matplotlib scikit-image
   ```
4. To deactivate the environment:
   ```
   deactivate
   ```

##### Mac/Linux:
1. Create a virtual environment:
   ```
   python3 -m venv myenv
   ```
2. Activate the virtual environment:
   ```
   source myenv/bin/activate
   ```
3. Install dependencies inside the environment:
   ```
   pip install opencv-python matplotlib scikit-image
   ```
4. To deactivate the environment:
   ```
   deactivate
   ```

#### Notes:
- Always activate your virtual environment before working on your project.
- Use a `requirements.txt` file to save your dependencies:
  ```
  pip freeze > requirements.txt
  ```
- To install dependencies from a `requirements.txt` file:
  ```
  pip install -r requirements.txt
  ```

````python
# Import necessary libraries
import cv2
import matplotlib
import matplotlib.pyplot as plt
from skimage import io

# Display library versions
print("OpenCV version:", cv2.__version__)
print("Matplotlib version:", matplotlib.__version__)
````

### 1. Load and Display an Image

````python
import os

file_path = '/Users/sridhar/projects/learnings/curriculum/sample.jpg'
print("File exists:", os.path.exists(file_path))

# Load the image using OpenCV (sample_image.jpg should exist in your directory)
img = cv2.imread(file_path)

# Check if the image is loaded successfully
if img is not None:
    # Convert from BGR to RGB for Matplotlib
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Display the image using Matplotlib
    plt.imshow(img_rgb)
    plt.title("Sample Image (RGB)")
    plt.axis("off")
    plt.show()
else:
    print("Error: Unable to load the image. Ensure the file path is correct.")
````

### 2. Convert Image to Grayscale

````python
if img is not None:
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Display the grayscale image
    plt.imshow(gray_img, cmap='gray')
    plt.title("Grayscale Image")
    plt.axis("off")
    plt.show()
else:
    print("Error: Image not loaded; skipping grayscale conversion.")
````

### 3. Extract Basic Properties of the Image

````python
if img is not None:
    print("Original Image Shape:", img.shape)  # Height, Width, Channels
    print("Grayscale Image Shape:", gray_img.shape)  # Height, Width
    print("Number of Pixels in Original Image:", img.size)
else:
    print("Error: Image not loaded; skipping property extraction.")
````

### 4. Read and Display Video Frames

````python
# Load a sample video (sample_video.mp4 should exist in your directory)
video_path = 'sample_video.mp4'
cap = cv2.VideoCapture(video_path)

if cap.isOpened():
    ret, frame = cap.read()
    if ret:
        # Convert BGR to RGB for display
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        # Display the first frame
        plt.imshow(frame_rgb)
        plt.title("First Video Frame (RGB)")
        plt.axis("off")
        plt.show()
    else:
        print("Error: Unable to read video frames.")
    cap.release()
else:
    print(f"Error: Unable to open video file '{video_path}'. Ensure the file path is correct.")
````
